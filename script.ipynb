```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def sawtooth_wave(t, amplitude=1, period=1):
    return amplitude * np.mod(t, period) / period

# 
t = np.linspace(0, 10, 1000)  # 0101000

# 
sawtooth = sawtooth_wave(t)

# 
plt.figure(figsize=(10, 5))

# 
plt.plot(t, sawtooth, label='Sawtooth Wave')

# 
plt.title('Sawtooth Wave Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# 
plt.legend()

# 
plt.grid(True)

# 
plt.show()
```

```python
sawtooth[:] = sawtooth + np.random.normal(0, 0.5)

# 
plt.figure(figsize=(10, 5))

# 
plt.plot(t, sawtooth, label='Sawtooth Wave')

# 
plt.title('Sawtooth Wave Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# 
plt.legend()

# 
plt.grid(True)

# 
plt.show()
```
