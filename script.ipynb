```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def sawtooth_wave(t, amplitude=1, period=1):
    return amplitude * np.mod(t, period) / period

# 定义时间范围
t = np.linspace(0, 10, 1000)  # 从0到10，共1000个点

# 计算锯齿波的值
sawtooth = sawtooth_wave(t)

# 创建图像
plt.figure(figsize=(10, 5))

# 绘制锯齿波形
plt.plot(t, sawtooth, label='Sawtooth Wave')

# 添加标题和标签
plt.title('Sawtooth Wave Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# 显示图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图像
plt.show()
```

```python
sawtooth[:] = sawtooth + np.random.normal(0, 0.5)

# 创建图像
plt.figure(figsize=(10, 5))

# 绘制锯齿波形
plt.plot(t, sawtooth, label='Sawtooth Wave')

# 添加标题和标签
plt.title('Sawtooth Wave Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# 显示图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图像
plt.show()
```
